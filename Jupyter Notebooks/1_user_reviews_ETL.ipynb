#### Extracción y exploración

```python
import pandas as pd
import ast
import tools
import warnings
warnings.filterwarnings("ignore")


ruta_items = "australian_user_reviews.json"

# Se lee de cada línea del dataset
filas_items = []
with open(ruta_items, "r", encoding="utf-8") as f:
    for line in f.readlines():
        filas_items.append(ast.literal_eval(line))

# Se convierte en dataframe
df_user_reviews = pd.DataFrame(filas_items)
df_user_reviews
```

```python
#Verificamos el tipo de datos que contiene
tools.ver_tipo_datos(df_user_reviews)
```

Observamos que no hay datos nulos, pero la columna reviews es una lista, lo más seguro es que tengamos que desanidarla. 

```python
filas_duplicadas = tools.duplicados(df_user_reviews, "user_id")
filas_duplicadas 
```

Podemos apreciar que hay 623 filas repetidas para la columna "user_id". Vamos a revisar si los datos en "reviews" dentro de la lista son iguales o contiene comentarios diferentes.   

```python
user_reviews = filas_duplicadas[filas_duplicadas["user_id"] == "yolofaceguy"]["reviews"]

for i in user_reviews:
    for r in i:
        print(r["review"])
    print("")


```

Podemos ver que los comentarios son exactamente iguales. Por lo tanto, decidimos de quedarnos con el primero y eliminar el resto. 

```python
df_user_reviews = df_user_reviews.drop_duplicates(subset="user_id", keep="first")
tools.duplicados(df_user_reviews, "user_id")
```

Analizaremos la columna "reviews" para entender más a fondo la estructura de los datos.

```python
df_user_reviews["reviews"][3]
```

Este dataFrame contiene 3 columnas y 25799 filas, sin valores nulos. Las columnas son:

* user_id: identificador único para el usuario.
* user_url: la url del perfil del usuario en streamcommunity.
* reviews: una lista de diccionarios. Para cada usuario se tiene uno o más            diccionario con el review. Cada diccionario contiene:
    * funny: indica si alguien puso emoticón de gracioso al review.
    * posted: fecha de posteo del review en formato "Posted April 21, 2011".
    * last_edited: fecha de la última edición.
    * item_id: identificador único del item, es decir, del juego.
    * helpful: estadística donde otros usuarios indican si fue útil la información.
    * recommend: booleano que indica si el usuario recomienda o no el juego.
    * review: sentencia string con los comentarios sobre el juego.

La columna "reviews" se presenta anidada, siendo una lista con uno o más diccionarios como elementos. Buscaremos generar una columna por cada diccionario para posteriormente hacer un registro por cada diccionario.



```python
df_desanidado = pd.json_normalize(df_user_reviews["reviews"])
df_desanidado
```

Observamos que al realizar esta transformación, perdemos el "user_id" y "user_url" al que pertenece cada diccionario, pero aún mantiene la misma posición. Para volver a tener la trazabilidad del usuario, lo concatenaremos con el dataframe anterior.

```python
df_desanidado = pd.concat([df_user_reviews[["user_id", "user_url"]], df_desanidado], axis=1)
df_desanidado
```

Al tener los diccionarios por columnas, generaremos un registro por cada diccionario con el usuario que genera dicha información.

```python
df_desanidado = pd.melt(df_desanidado, id_vars=["user_id", "user_url"], 
                       value_vars=list(range(9)),
                       value_name="reviews")
df_desanidado.head()
```

Al hacer lo últimos, observamos que se presentan registros None. Esto ocurre porque hay usuarios que hicieron más reviews que otros. 
En este ejemplo se puede ver este caso:

```python
df_desanidado[df_desanidado["user_id"]=="76561197970982479"]
```

Eliminamos los registros que tienen None en "reviews".

```python
df_desanidado = df_desanidado.dropna()
# Verificamos que solo queden el 'user_id' con la cantidad de diccionarios que le corresponde
df_desanidado[df_desanidado["user_id"]=="76561197970982479"]
```

En este punto, ya podemos convertir cada diccionario en columna.

```python
# Separamos por columnas cada una de las claves de "reviews"
df_user_reviews = df_desanidado["reviews"].apply(pd.Series, dtype="object")
df_user_reviews = df_user_reviews.add_prefix("reviews_")
df_user_reviews.head()
```

Tras este proceso, podemos ver que la columna de "user_id" y "user_url" se perdió nuevamente, por lo que volveremos a concatenarlos.

```python
df_user_reviews = pd.concat([df_desanidado[["user_id", "user_url"]], df_user_reviews], axis=1)
df_user_reviews.head()
```

Podemos visualizar que hay valores faltantes en algunas columnas, pero no están como nulos, probablemente porque deben tener un espacio. Lo comprobaremos y si es así, los rellenaremos como valores nulos.

```python
df_user_reviews["reviews_last_edited"][0]
```

```python
df_user_reviews.replace("", None, inplace=True)
df_user_reviews
```

Analizamos el tipo de datos y los nulos que quedaron luego de desanidar la columna "reviews".

```python
tools.ver_tipo_datos(df_user_reviews)
```

Hay un 86 y 89% de faltantes de datos en las columnas "reviews_funny" y "reviews_last_edited" respectivamente, por lo que decidimos eliminar estas columnas.

```python
df_user_reviews = df_user_reviews.drop(columns=["reviews_funny", "reviews_last_edited"])
df_user_reviews.columns
```

#### Columna "reviews_posted"
La fecha se encuentra en este formato Posted April 17, 2015, pero necesitamos que la fecha esté en formato YYYY-MM-DD. Por lo tanto, procesaremos la fecha y extraeremos los elementos relevantes.

```python

# Nos aseguramos de que "posted" sea de tipo datetime
df_user_reviews["reviews_posted"] = pd.to_datetime(df_user_reviews["reviews_posted"].astype(str).str.replace(r'Posted |,|\.', '', regex=True), errors='coerce')

# Creamos la columna "year" a partir de "posted"
df_user_reviews["year"] = df_user_reviews["reviews_posted"].dt.year.astype("Int64")

# Ordenamos el DataFrame por "item_id" y "year" para asegurar que la interpolación se haga correctamente
df_user_reviews = df_user_reviews.sort_values(["user_id", "year"])


df_user_reviews



```

```python
# Rellenamos valores nulos en "year" mediante interpolación lineal por grupo (item_id)
df_user_reviews['year'] = df_user_reviews.groupby('reviews_item_id', group_keys=False)['year'].apply(lambda group: group.interpolate(method='pad') if group.notna().any() else group)

# Si aún observamos valores nulos después de la interpolación, los rellenamos con la mediana. 
df_user_reviews['year'] = df_user_reviews['year'].fillna(df_user_reviews['year'].median())

df_user_reviews
```

Eliminaremos la columna "reviews_posted" porque no aporta mayor información, ya que solo indica el día y el mes del posteo.

```python
df_user_reviews = df_user_reviews.drop(["user_url", "reviews_posted"], axis=1)
df_user_reviews
```

#### Columna "reviews_review"

```python
df_user_reviews = df_user_reviews.dropna(subset=["reviews_review"])
# Verificamos nulos
tools.ver_tipo_datos(df_user_reviews)
```
