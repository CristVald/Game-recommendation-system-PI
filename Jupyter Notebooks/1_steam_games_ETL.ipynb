# ETL del dataset "steam_games.json.gz"

##### En este archivo de iniciaremos con la Extracción, transformación y carga de los datos. El objetivo es extraer el contenido de un JSON comprimido para visualizar los datos a través de un dataFrame de Pandas y luego guardarlos en un formato CSV.   



```python
# Importamos las librerías a utilizar.
import json
import pandas as pd
import ast
import gzip
from tqdm import tqdm
import tools
import numpy as np

# Creamos la función para leer las líneas de cada dataset.
def extraer_datos(ruta, tipo="json"):
    filas = []  
    with gzip.open(ruta, "rt", encoding="utf-8") as f:
        for line in tqdm(f.readlines(), desc=f"Leyendo {ruta}"):
            if tipo == "json":
                data = json.loads(line)
            elif tipo == "literal":
                data = ast.literal_eval(line)
            filas.append(data)
    return pd.DataFrame(filas)

# Convertimos en dataframe el dataset
df_steam_games = extraer_datos("steam_games.json.gz", tipo="json")
df_steam_games
```

```python
# Calculamos el porcentaje de datos faltantes por columna
tools.porcentaje_faltante(df_steam_games)
```

Observamos que el dataFrame contiene una gran cantidad de datos nulos, por lo que decidimos eliminarlos.

```python
df_steam_games = df_steam_games.dropna(how="all").reset_index(drop=True)
df_steam_games
```

```python
# Verificaremos los tipos de datos en el dataFrame
tools.ver_tipo_datos(df_steam_games)
```

Al analizar el contenido de las columnas, vamos a asegurarnos que todas las columnas que tengan un valor nulo o vacio, pasen a tener el valor de NaN en el dataFrame

```python
df_steam_games.replace(["","None","Null"], np.nan, inplace = True)
```

```python
# Renombraremos las columnas

df_steam_games.rename(columns={"app_name": "name", "id": "item_id"}, inplace=True)
```

Compararemos las columnas "Publisher" y "Developer" para ver qué cantidad de valores repetidos o porcentaje de NaN contienen. 

```python
df_steam_games[["publisher", "developer"]].isnull().sum()
```

```python
tools.duplicados(df_steam_games, "item_id")
```

Como podemos apreciar, la salida de la función utilizada nos indica que existen 2 ID´s  con el mismo nombre en la base de datos. Aunque, el segundo par de ID´s tiene por nombre NaN. Entonces, vamos a compararlas ver qué información relevante nos pueden dar o si podemos eliminarla, porque la información relevante ya se encuentra en otro registro del dataset.

```python
#Vamos a buscar por la parte diferente, es decir en la columna "developer" 
df_steam_games[df_steam_games["developer"]=="Rocksteady Studios,Feral Interactive (Mac)"]
```

Podemos ver que el registro en la columna developer, ya está registrado otras 3 veces y se encuentra completo, por lo que tomaremos la decisión de eliminar los dos registros con ID igual a NaN.

```python
#eliminaremos los indices repetidos o no deseados
indices_a_eliminar = [14573, 4, 30961]
df_steam_games = df_steam_games.drop(indices_a_eliminar)

#nuevamente, verificaremos si hay duplicados
tools.duplicados(df_steam_games,"item_id")
```

```python
df_steam_games
```

#### Nos enfocaremos en la columna "release_date"
Vamos a extraer el año de esta columna, ya que es la parte que más nos importa. Luego, pondremos la información extraída en una nueva columna para verificar sus valores. Por último, eliminaremos la columna "release_date".  

```python
df_steam_games["release_date"].value_counts()
```

```python
# Creamos una columna para el año
df_steam_games["release_year"] = df_steam_games["release_date"].apply(tools.separar_año)

# eliminamos la columna "release_date"
df_steam_games = df_steam_games.drop("release_date", axis=1)
df_steam_games.head(5)
```

```python
df_steam_games["release_year"].unique()
```

#### Columna "price"


```python
df_steam_games["price"].unique()
```

Podemos observar que la columna "Price" contiene principalmente valores flotantes, pero también hay valores de texto o nulos que reemplazaremos por un cero. 

```python
df_steam_games["price"] = df_steam_games["price"].apply(tools.reemplazar_flotante)
df_steam_games["price"].dtype
```

#### Columnas "publisher", "app_name", "title", "developer"

Observamos muchos tipos de datos nulos, vamos a transformarlos para que el dato nulo sea uniforme. 

```python
columnas_a_completar = ["publisher", "name", "title", "developer"]
# Rellenamos los nulos
df = df_steam_games[columnas_a_completar].fillna("Sin dato disponible")
# Borramos las columnas originales y concatenamos las columnas rellenas con todo el dataframe
df_steam_games = pd.concat([df_steam_games.drop(columnas_a_completar, axis=1), df], axis=1)
df_steam_games.head()
```

#### Columna "genres"

La columna "genres" se compone por una lista de los distintos géneros de videojuegos. Se necesita hacer una fila con cada género.

```python
df_steam_games = df_steam_games.explode("genres")
df_steam_games = df_steam_games.dropna(subset=["genres"])
df_steam_games.head(20)
```

#### Columnas "tags", "specs", "url" y "reviews_url"

Eliminaremos estas columnas, porque no las utilizaremos para las consultas en la APi ni para el modelo de recomendación. 

```python
df_steam_games = df_steam_games.drop(["tags", "specs", "url", "reviews_url"], axis=1)
df_steam_games.columns
```

```python
# Verificamos el tipo de datos y si hay nulos
tools.ver_tipo_datos(df_steam_games)
```

#### Guardamos el dataset

```python
archivo = "steam_games_cleaned.csv"
df_steam_games.to_csv(archivo, index=False, encoding="utf-8")
print(f"Se guardó el archivo {archivo}")
```
